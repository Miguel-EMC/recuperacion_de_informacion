Lista los archivos de la carpeta y carga el contenido de cada uno en un diccionario.

```python
import time
query = 'amor'
start_time = time.time()
result = []

for key in contenidos.keys():
    if query in contenidos[key]:
        print(f'La palabra "{query}" existe en el contenido.')
        result.append(key)
    else:
        print(f'La palabra "{query}" no existe.')

end_time = time.time()
print(f'Los archivos que contienen la palabra "{query}" son: {result}')
print(f'Tiempo de búsqueda: {end_time - start_time} segundos')
```

```python
def search(query, contenidos):
    resultados = []
    for key in contenidos.keys():
        if query in contenidos[key]:
            resultados.append(key)
    return resultados
```
